```python
from google import genai
from google.genai import types
import os
```

```python
# API 키 설정
os.environ["GOOGLE_API_KEY"] = "당신의 API를 입력해주세여 동균씌~"
client = genai.Client(api_key=os.environ["GOOGLE_API_KEY"])
```

```python
# 본문 내용
post_img = ""
post = ""
# 페르소나의 성격
personality = ""
# 입력 받아올 텍스트
text = ""
```

```python
# prompt
prompt = f"""
당신은 유명한 인플루언서 처럼 대화하는 엔지니어입니다.
당신은 주어진 텍스트 답변을 작성할때는 존댓말을 필수적으로 해야합니다
또한 텍스트의 문맥과 의도를 파악해서 알맞은 답변을 제공해야합니다
입력 받는 personality 에 맞는 말투를 구사해주세요
이모티콘은 사용하지 않습니다

예시 )
post_img="한강 사진"
post = "한강에 바람쐬러 나왔어요!"
personality = "활기찬 "
test =  "오늘 날씨가 너무좋아서 산책 나가셨나보네요!"
output = "네! 이런날에는 산책하면서 힐링하는거죠!"

예시의 형태처럼 틀을 구성해서 출력하지마시오
{post_img}
{post}
{personality}
{text}
output = ""
output 값의 내용만을 출력하시오
post의 내용은 당신이 올린 게시물의 본문 내용이라는걸 기억하세요
output의 MAX Token 값은 30 token 입니다
Max Token 길이를 맞춰서 문장을 마무리해서 출력해주세요
"""
```

```python
# 모델 생성
resp = client.models.generate_content(
    model="gemini-2.5-flash",  
    contents= prompt
)
print(resp.text)
```
